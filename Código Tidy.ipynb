```python
import pandas as pd
import numpy as np
```

```python
url= "https://docs.google.com/spreadsheets/d/e/2PACX-1vRkZgcI04-ftSrDHF45igtAisHZ5z85aIYAbi5eQhCeo_r4AuiCTfQR9fbfJ0BewzZKDdgbrgG1dLV-/pub?output=xlsx"
base=pd.read_excel(url,index_col=1)
base.head()
```

```python
base = base.rename(columns={'¿Cuál es el rango de precios promedio por persona en su establecimiento (plato fuerte + bebida)?':'Rango de Precios',
                           '¿Cómo espera que la construcción del metro y la transformación en un nodo multimodal impacte la afluencia de clientes a su negocio?':'Afluencia',
                           'Desde que inició la construcción del metro en la Calle 72, ¿qué cambios ha percibido en el entorno físico del sector (espacio público, ruido, acceso, visibilidad del local)?'
                            :'Cambios Físicos',
                            '¿Cómo espera que cambien los ingresos de su establecimiento una vez esté en funcionamiento la estación multimodal de la Calle 72?':
                            'Ingresos',
                            '¿Considera que el nuevo nodo de transporte atraerá un tipo diferente de clientela a su establecimiento?':'Clientela',
                            'En su opinión, ¿Cuáles son los principales riesgos o preocupaciones frente a la entrada en operación del metro en la Calle 72?':
                            'Riesgos',
                            '¿Cuál(es) medida(s) considera usted que su negocio debería adoptar para adaptarse a los cambios generados por las obras del metro en la zona?':
                            'Medidas de Adaptación'})
base.drop('Marca temporal', axis=1, inplace=True)
base.head()
```

```python
base.index.name='Identificador'
```

```python
base['Rango de Precios'] = base['Rango de Precios'].replace({
    'Menos de $25.000': 0,
    'Entre $25.000 y $40.000': 1,
    'Más de $40.000': 2
}).astype('int')

base['Afluencia'] = base['Afluencia'].replace({
   'Aumentará significativamente':0,'Aumentará levemente':1, 'No cambiará':2, 'Disminuirá levemente':3,
    'Disminuirá significativamente':4, 'No sabe / Prefiere no responder':5
}).astype('int')

base['Ingresos'] = base['Ingresos'].replace({
    'Aumentarán significativamente':0,'Aumentarán levemente':1,'No cambiarán':2,'Disminuirán levemente':3,
    'Disminuirán significativamente':4,'No sabe / Prefiere no responder':5
}).astype('int')

base['Clientela'] = base['Clientela'].replace({
    'Sí, atraerá más clientes ocasionales o de paso':0,
    'Sí, atraerá más trabajadores o personas que transitan por motivos laborales':1,
    'No, la clientela se mantendrá similar':2, 'No lo sabe / Prefiere no responder':3
}).astype('int')

base.head()
```

```python
opciones = [
    "Mayor dificultad de acceso vehicular o peatonal",
    "Aumento del ruido y el polvo en la zona",
    "Menor visibilidad o tránsito frente al local",
    "Mejora o ampliación de aceras y andenes",
    "No ha percibido cambios relevantes",
    "Prefiere no responder"
]

opt_to_cf = {opt: f"cf{i}" for i, opt in enumerate(opciones, start=1)}

listas = (
    base['Cambios Físicos']
      .fillna('')               
      .str.split(',')           
      .apply(lambda lst: [s.strip() for s in lst])
)

for opt, cf_col in opt_to_cf.items():
    base[cf_col] = listas.apply(lambda respuestas: int(opt in respuestas))

base.drop(columns=['Cambios Físicos'], inplace=True)

base.head()
```

```python
opciones_1 = [
    "Aumento de la competencia comercial en la zona",
    "Cambios en el perfil socioeconómico de los visitantes",
    "Pérdida de clientes habituales por cambios en el entorno",
    "Mayor congestión o desorden en el espacio público",
    "No percibe riesgos importantes",
    "Prefiere no responder"
]

opt_to_r = {opt: f"r{i}" for i, opt in enumerate(opciones_1, start=1)}

listas_1 = (
    base['Riesgos']
      .fillna('')
      .str.split(',')
      .apply(lambda lst: [s.strip() for s in lst])
)

for opt, r_col in opt_to_r.items():
    base[r_col] = listas_1.apply(lambda respuestas: int(opt in respuestas))

base.drop(columns=['Riesgos'], inplace=True)

base.head()
```

```python
opciones_2 = [
     "Ajustar los horarios de apertura o cierre del establecimiento",
    "Modificar el menú o los productos ofrecidos",
    "Cambiar la forma de atención al cliente (por ejemplo, servicio en mesa, autoservicio, etc.)",
    "Invertir en mejorar la visibilidad del local (fachada, señalización, iluminación)",
    "Implementar canales de venta digitales o servicios a domicilio",
    "Diversificar la oferta de productos o servicios para atraer nuevos clientes",
    "Modificar precios o lanzar promociones para mantener o aumentar la demanda",
    "Fortalecer la presencia en redes sociales o mejorar las estrategias de marketing",
    "Buscar apoyo institucional o establecer alianzas con otros negocios del sector",
    "Aún no se ha implementado ninguna medida, pero estamos considerando hacerlo",
    "No creemos que sea necesario adoptar ninguna medida",
    "Prefiero no responder"
]

# Mapeo a med1…med12
opt_to_med = {opt: f"med{i}" for i, opt in enumerate(opciones_2, start=1)}

# Partir y limpiar respuestas de la columna correcta
listas_2 = (
    base['Medidas de Adaptación']
      .fillna('')
      .str.split(',')
      .apply(lambda lst: [s.strip() for s in lst])
)

# Crear med1…med12
for opt, med_col in opt_to_med.items():
    base[med_col] = listas_2.apply(lambda respuestas: int(opt in respuestas))

# Eliminar la columna original
base.drop(columns=['Medidas de Adaptación'], inplace=True)

# Verificación final
base.tail()
```

```python
base.to_excel('data_tidy.xlsx', index=True)
```
